```python
import pandas as pd
```

```python
df1_original = pd.read_csv("data/raw/Dataset_SCO_KVM_MONS_GRC_IZD_DMA_MTK_20250805.csv")
df2_original = pd.read_csv("data/raw/Etagi_secondary_classified_dataset_20250805.csv")
df3_original = pd.read_csv("data/raw/Etagi_secondary_dataset_20250805.csv")
```

```python
df1 = df1_original.copy()
df2 = df2_original.copy()
df3 = df3_original.copy()
```

```python
print(f"df1: {df1.shape[0]} x {df1.shape[1]}")
print(f"df2: {df2.shape[0]} x {df2.shape[1]}")
print(f"df3: {df3.shape[0]} x {df3.shape[1]}")
```

```python
pd.set_option('display.max_columns', None)
```

```python
def clean_dfs(dfs):    
    df1, df2, df3 = dfs

    dfs = [df.drop_duplicates() for df in dfs]

    dfs = [
        df.replace("", pd.NA).dropna(how="all").dropna(axis=1, how="all") for df in dfs
    ]

    df1, df2, df3 = dfs

    df1 = df1.drop(columns=['ID помещения', 'BIM идентификатор CRM', 'Клиент'])
    df2 = df2.drop(columns=['object_id', 'url', 'parent_id', 'duplicates_url', 'city_id'])
    df3 = df3.drop(columns=['source_city_id', 'city_id', 'newcomplex_id', 'newhouse_id', 'object_id'])

    df1 = df1[df1['Вид помещения'] == 'Квартира']

    df1 = df1.drop(columns=['Вид помещения'])

    dfs = [df1, df2, df3]
    return dfs
```

```python
dfs = [df1, df2, df3]
df1_cleared, df2_cleared, df3_cleared = clean_dfs(dfs)
```
